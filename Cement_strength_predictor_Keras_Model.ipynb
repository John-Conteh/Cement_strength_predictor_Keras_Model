## Cement strength predictor model 

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras 
from keras.models import Sequential 
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD, Adam
```

```python
#file_url = "http://storage.googleapis.com/download.tensorflow.org/data/heart.csv"
#df = pd.read_csv('../DATA/moviereviews.csv')

df = pd.read_csv('../My_DATASET/concrete_data.csv')
```

```python
X = df.drop('Strength',axis=1)
y = df['Strength']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
scaled_X_train.shape
```

```python
model = Sequential()
model.add(Dense(200, input_dim=8, activation = 'relu'))
model.add(Dense(200, input_dim=200, activation = 'relu'))
model.add(Dense(200, input_dim=200, activation = 'relu'))
model.add(Dense(1, activation = 'linear'))
```

```python
keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.99, amsgrad=False)
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])

#model.compile(optimizer='sgd',loss='mse',metrics=[tf.keras.metrics.RootMeanSquaredError()])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=300, batch_size=20, validation_split=0.15, validation_data=None,verbose=1)
```

```python
keras.backend.clear_session()
```

```python
pred_Strength = model.predict(X_test)
```

```python
pred_Strength
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,pred_Strength)
MSE = mean_squared_error(y_test,pred_Strength)
RMSE = np.sqrt(MSE)
```
